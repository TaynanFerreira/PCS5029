```python
import pandas as pd
import csv
import matplotlib.pyplot as plt
```

```python
headers = ['Wall Time','Step','Accuracy']
```

## Accuracy: variando tamanho Kernel

```python
df5 = pd.read_csv('./logs_v5/run_training-tag-acc.csv',names=headers)
df9 = pd.read_csv('./logs_v9/run_training-tag-acc.csv',names=headers)
df10 = pd.read_csv('./logs_v10/run_training-tag-acc.csv',names=headers)
```

```python
y5 = df5['Accuracy']
y9 = df9['Accuracy']
y10 = df10['Accuracy']

plt.plot(y5)
plt.plot(y9)
plt.plot(y10)

plt.legend(['f = 3', 'f = 4', 'f = 5'], loc='upper left')

plt.show()
```
